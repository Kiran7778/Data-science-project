```python
Q1 = df['Year'].quantile(0.25)
Q3 = df['Year'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Year'] < lower_bound) | (df['Year'] > upper_bound)]
print(outliers)
```

This code finds outliers in the Year column of the DataFrame df. It calculates the interquartile range (IQR) and then determines which values fall outside the range of 1.5 times the IQR below the first quartile (Q1) and above the third quartile (Q3). The outliers are then printed.
